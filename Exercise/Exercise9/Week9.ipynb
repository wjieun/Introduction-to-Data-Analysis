#California housing data 분석 및 코드 예시

- 위에서 설명한 캘리포니아 데이터 프레임을 불러오고 분석해봅시다.
- 아래의 코드 예시를 실행시켜보고 출력을 확인해봅시다.

```
data_path = 'sample_data/california_housing_train.csv'
df = pd.read_csv(data_path)
print(df.describe())

# 주택 나이에 대한 히스토그램 분포
plt.hist(df['housing_median_age'])
plt.title("hist_housing_median_age")
plt.show()
print()

# x를 경도, y를 위도로 하는 선 그래프
plt.title("plot_x_longitude_y_latitude")
plt.plot(df['longitude'],df['latitude'])
plt.show()
print()

# 주택 가치에 대한 박스 그래프
plt.title("box_median_house_value")
plt.boxplot(df['median_house_value'])
plt.show()
```

#문제3(Cont.)

- plt에는 colorbar()라는 함수가 존재하는데, 이를 사용해 어떤 색이 어떤 값을 나타내는지 쉽게 볼 수 있습니다.
- scatter에 color를 지정하고 colorbar 함수가 어떤 역할을 하는지 알아 봅시다.
- scatter plot에서 color는 "colors = " 대신 "c = "을 사용하고, 값이 들어갈 수 있습니다.
- 값을 사용하게 될 경우 각 점에 대한 가중치 값 표현이 가능합니다.
- 아래의 예시를 직접 출력해보시기 바랍니다.

```
# X 위치
loc_x = [0, 0.9, 1.8, 2.7, 3.6, 4.5, 5.4, 6.3, 7.2, 8.1, 9]
# Y 위치
loc_y = [18, 19.1, 19.3, 18.7, 17.2, 14.8, 11.5, 7.3, 2.3, -3.7, -10.5]
# Y 방향 속도
velocity_y = [5.2, 2.3, -0.7, -3.6, -6.6, -9.5, -12.4, -15.4, -18.3, -21.3, -24.2]

plt.scatter(loc_x, loc_y, c = velocity_y)
plt.title("Projectile motion")
plt.xlabel("X_location")
plt.ylabel("Y_location")
plt.colorbar()
plt.show()
```

#문제3(Cont.)

- 앞서 실행해보신 것처럼, Color bar를 사용하여 어떤 색이 어떤 값을 표현하는지 그라데이션으로 명확히 보실 수 있습니다.
- 이를 사용해 문제 3의 scatter plot에 주택 가치를 색으로 주어 colorbar를 출력해서 어느 지역에 높은 집 값이 분포하는지 알아봅시다.

```
# 문제 3에 colorbar를 적용해 출력해보는 코드를 여기에 작성하세요.
plt.scatter(df['longitude'], df['latitude'], c=df['median_house_value'])
plt.colorbar()
plt.show()
```
